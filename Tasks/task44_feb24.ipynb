## BAR GRAPH 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np 
```

```python
x=['sports','companies','tv','shops','theaters','cities','villages']
y=[23,24,21,100,1000,34,45]
y1=[22,21,25,48,50,2000]
```

```python

plt.bar(x,y,color='g',width=0.5)
plt.xlabel('categorical data ')
plt.ylabel('count')
plt.title('BAR GRAPH ')
plt.show()
```

```python
year=[2015,2016,2017,2020,2021]
x=[23,24,25,26,27]
y=[12,13,18,20,34]
w=0.5
l=np.arange(5)


plt.bar(l,x,color='c',width=w,label='CSE')
plt.bar(l+w,y,color='k',width=w,label='ECE')
plt.legend()

plt.xticks(ticks=l+0.2,labels=year)
plt.show()
```

## pie charts 

```python
data=[23,24,34,21,21]
plt.pie(data,labels=['ind','aus','newzwland','switzerland','usa'],colors=['red','green','cyan','black','m'],startangle=90,shadow=True)
plt.show()
```

## LINE CHARTS

```python
x=[11,12,13,14,15]
y=[13,14,13,12,10]

plt.plot(x,y,marker='d',ms=10,mec='g',mfc='c',linestyle='-',linewidth=3)

plt.show()
```

## HISTOGRAM

```python
data=[23,25,70,90,100,32,87,44]
intervals=[0,25,50,100]
```

```python
plt.hist(data,facecolor='c',bins=intervals)
plt.xticks(intervals)
plt.show()
```
